# Correlation Matrix of Electoral Bond Transactions and Political Events

```python
# Merge datasets on 'Year'
purchase_encash = pd.merge(purchase_year_plot, encashment_year_plot, on='Year', how='outer')
purchase_encash_elec = pd.merge(purchase_encash, elections_year_plot, on='Year', how='outer')

correlation_matrix = purchase_encash_elec.corr()

fig = plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

plt.title('Correlation Matrix of Electoral Bond Transactions and Political Events')
plt.show()
fig.savefig('Correlation Matrix.jpg')
```

*The correlation coefficients provide a quantitative measure of the linear relationship between the variables. A coefficient close to 1 indicates a strong positive correlation, while a coefficient close to -1 indicates a strong negative correlation. A coefficient around 0 suggests no linear correlation.*

**Strong Positive Correlation between Encashment and Number of Elections:**

**There is a strong positive correlation of approximately 0.95 between the amount encashed and the number of elections held.**

**Insight:**

**This indicates that as the number of elections held increases, there is a corresponding increase in the amount of electoral bonds encashed. It suggests that electoral bond transactions might be influenced by election cycles and political activities.**

**Moderate Positive Correlation between Purchase and Number of Elections:**

**There is a moderate positive correlation of approximately 0.89 between the amount purchased and the number of elections held.**

**Insight:**

**This suggests that there is a tendency for higher electoral bond purchases during years with more elections. It could be attributed to increased political spending during election periods.**

# Top 3 Purchases every year

```python
# Copy the original purchase_data dataframe
top_purchase_year = purchase_data.copy()

# Convert purchase amounts to crores
top_purchase_year['purchase'] = top_purchase_year['purchase'].apply(format_crore)

# Group by Year and purchaser_name, then sum the purchase amounts
top_purchase_year = top_purchase_year.groupby(['Year', 'purchaser_name'])['purchase'].sum().reset_index()

# Rank the purchasers based on their total purchase amounts in each year
top_purchase_year['rank'] =  top_purchase_year.groupby(['Year'])['purchase'].rank(method='dense', ascending=False)

# Filter out only the top 3 purchasers in each year
top_purchase_year = top_purchase_year.loc[top_purchase_year['rank']<=3].sort_values(['Year','rank'])

# Copy the purchase_year_plot dataframe
yearly_purchase = purchase_year_plot.copy()

# Rename purchase column
yearly_purchase.rename(columns={'purchase': 'yearly_purchase'}, inplace=True)

# Merge datasets on Year
top_purchase_year = pd.merge(top_purchase_year, yearly_purchase, on='Year', how='outer')

# Calculate the percentage of each purchase relative to the total purchases in each year
top_purchase_year['percentage'] = round((top_purchase_year['purchase']/top_purchase_year['yearly_purchase'])*100,2)

# Create a label combining purchaser_name and Year
top_purchase_year['label'] = top_purchase_year['purchaser_name'] + ' | ' + top_purchase_year['Year'].astype('str')

top_purchase_year.head()
```

```python
fig = plt.figure(figsize=(20, 8))

ax = sns.barplot(data=top_purchase_year, x='purchase', y='label', hue='purchaser_name')

# Loop through each bar patch and set edgecolor
for patch in ax.patches:
    patch.set_edgecolor('black')

# Add bar labels
for p in ax.patches:
    width = p.get_width()
    if width > 0:
        ax.annotate(f'{width} cr', 
                    xy=(width, p.get_y() + p.get_height() / 2), 
                    xytext=(5, 0), 
                    textcoords='offset points', 
                    ha='left', 
                    va='center')

plt.xlabel('Amount Purchased (in cr)')
plt.ylabel('Purchaser')
plt.title('Top 3 Purchasers - Yearly')

ax.legend_.remove()

plt.tight_layout()
plt.show()
fig.savefig('Top 3 Purchases.jpg')
```

**Top Purchasers Over the Years: The top purchasers of electoral bonds have varied across different years. In 2019, "KEVENTER FOODPARK INFRA LIMITED" was the top purchaser with a purchase amount of 195 crore, followed closely by "MADANLAL LTD." with 185.5 crore. However, in subsequent years, the top purchasers changed, with "FUTURE GAMING AND HOTEL SERVICES PR" emerging as the top purchaser in both 2021 and 2022, and "MEGHA ENGINEERING AND INFRASTRUCTURES LIMITED" leading the purchases in 2023.**

**Consistent Players: Some purchasers maintain their presence among the top ranks across multiple years. For example, "MEGHA ENGINEERING AND INFRASTRUCTURES LIMITED" consistently appears among the top purchasers in 2020, 2021, and 2023, indicating a sustained interest or requirement for electoral bonds.**

**Variability in Purchase Amounts: The purchase amounts vary significantly between different purchasers and years. For instance, the purchase amount by "FUTURE GAMING AND HOTEL SERVICES PR" increased by approximately 122.67% from 2020 to 2021 (150 crore to 334 crore) and further increased by about 21.86% from 2021 to 2022 (334 crore to 405 crore), showcasing a substantial rise in their electoral bond purchases over the years.**

**Yearly Fluctuations: There are fluctuations in the ranking and purchase amounts of purchasers from year to year. This indicates potential shifts in political or economic dynamics, changes in funding strategies, or alterations in business interests that influence the purchase of electoral bonds.**

**Effect of Regulatory Changes: Changes in regulations or business strategies may impact the purchase behavior of entities. For example, the decrease in purchase amount by "FUTURE GAMING AND HOTEL SERVICES PR" from 405 crore in 2022 to 256 crore in 2023 could be attributed to regulatory changes, market conditions, or internal factors affecting their decision-making process.**

**Emergence of New Players: The appearance of new purchasers such as "RUNGTA SONS P LTD" and "SHIRDI SAI ELECTRICALS LTD" in the top ranks in 2024 suggests the entry of new entities into the electoral bond market, potentially indicating diversification of funding sources or changes in business strategies among various companies.**
